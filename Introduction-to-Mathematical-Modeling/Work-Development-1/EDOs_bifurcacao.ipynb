Ponto de equilibrio da equação mais simples dx/dt = ax

```python
# Escolhemos valores de t por exemplo entre 0 e 100
t = np.linspace(0,100)

# cria vetor que vai armazenar solucoes
dxdt = np.zeros(np.size(t))

ep = 0 # variavel que armazena o ponto de equilibrio

# fazemos x ter valores negativos e positivos para comparar
x = -5

for y in range(0,10):
  # fazemos o parametro a conter valores negativos e positivos para comparar
  a = -24

  for i in range(0,np.size(t)):

    # calcula a funcao em cada ponto x
    dxdt[i] = a*x
    # ponto que intercepta eixo
    if (a*x == 0):
      ep = t[i]
    a = a+1 # avança o valor de a

  plt.plot(t,dxdt, '--')
  x = x+1 # avança o valor de x

# plota linha vertical no ponto de equilibrio - bifucacao
plt.axvline(x = ep, color = 'k', label = 'axvline - full height')
```

## Modelo Populacional Logistico

```python
# Exemplo para plotar grafico fx = ax(1-x) para a = 3.2

# define a funcao ax(1-x)
def fax(a,x):
  return a*x*(1-x)

# define intervalo de valores de a
A = np.arange(-1,1.25,0.25)
# define intervalo de valores de x
x = np.linspace(-10,10)

for a in A:
  fx = fax(a,x)
  plt.plot(x,fx)
  plt.grid()
```

```python
# plotando variacoes de a

for i in range(0,5):
  fx = fax(i,x)
  plt.plot(x,fx)
```

Inserindo termo de colheita constante:

```python
# Define f(x)
def fxh(x,h):
  return x*(1-x) - h

h = 1/4

# define intervalo para x
x = np.linspace(0,1)

fx = fxh(x,h)
plt.plot(x,fx)
```

```python
# plotando variacoes de h
plt.plot(x,fxh(x,1/4))
plt.plot(x,fxh(x,1/3))
plt.plot(x,fxh(x,1/6))
plt.plot(x,fax(0,x), 'k--')
plt.legend(['h=1/4', 'h>1/4', 'h<1/4'])
```
